# Семинар

```python
from sympy import *
```

```python
v = Function('v')
t = symbols('t')
m = 1
g = 1.6
```

```python
dsolve(diff(v(t), t) + g, v(t))
```

```python
vt = dsolve(diff(v(t), t) + g, v(t), ics={v(t).subs(t,0):100})
vt
```

```python
vt.rhs
```

```python
h = Function('h')
ht = dsolve(diff(h(t), t) - vt.rhs, h(t), ics={h(t).subs(t,0):0})
ht
```

```python
solve(vt.rhs, t, 0)
```

```python
ht.rhs.subs(t, 62.5)
```

```python
plot(ht.rhs, (t, 0, 150))
```

```python
plot(vt.rhs, (t, 0, 150))
```

# ДЗ

```python
k = 0.01
```

```python
dsolve(diff(v(t), t) + g + k * v(t), v(t))
```

```python
vt1 = dsolve(diff(v(t), t) + g + k * v(t), v(t), ics={v(t).subs(t,0):100})
vt1
```

```python
vt1.rhs
```

```python
ht1 = dsolve(diff(h(t), t) - vt1.rhs, h(t), ics={h(t).subs(t,0):0})
ht1
```

```python
t1 = solve(vt1.rhs, t, 0)
print('Время подъема:', t1[0][0])
```

```python
h1 = ht1.rhs.subs(t, t1[0][0])
print('Макс. высота:', h1)
```

```python
t2 = solve(ht1.rhs, t, 0)
t3 = t2[1][0] - t1[0][0]
print("Время возвращения:", t3)
```

```python
plot(ht1.rhs, (t, 0, 150))
```

```python
plot(vt1.rhs, (t, 0, 150))
```
