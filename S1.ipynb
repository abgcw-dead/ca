# Семинар

```python
from sympy import *
```

```python
v = Function('v')
t = symbols('t')
m = 1
g = 1.6
```

```python
dsolve(diff(v(t), t) + g, v(t))
```

```python
vt = dsolve(diff(v(t), t) + g, v(t), ics={v(t).subs(t,0):100})
vt
```

```python
vt.rhs
```

```python
h = Function('h')
ht = dsolve(diff(h(t), t) - vt.rhs, h(t), ics={h(t).subs(t,0):0})
ht
```

```python
solve(vt.rhs, t, 0)
```

```python
ht.rhs.subs(t, 62.5)
```

```python
plot(ht.rhs, (t, 0, 150))
```

```python
plot(vt.rhs, (t, 0, 150))
```

```python
k = 0.01
```

```python
dsolve(diff(v(t), t) + g + k * v(t), v(t))
```

```python
vt1.rhs
```

```python
ht1 = dsolve(diff(h(t), t) - vt1.rhs, h(t), ics={h(t).subs(t,0):0})
ht1
```

```python
ht1.rhs.subs(t, t1[0][0])
```
